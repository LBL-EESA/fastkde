The following code block generates samples from the Duong (2005) PDF sets multiple times, calculates the ISE and timing for the self-consistent PDF estimate of each draw, and plots the first PDF estimate for each.  Following this are some examples of running the test with different parameters, with discussion in between.

```python
%matplotlib inline

import selfConsistentDensityEstimate as sc
reload(sc)

#Import the dictionary of sumOfNormals objects and the timer
#Note this requires sciPy >= 0.14
from duongTests import duongTestDict

#import some Python core libraries
import copy
import time

#Import the numerical Python library
from numpy import *

#***********************
# Define a simple timer
#***********************
class Timer():
   def __enter__(self): 
     self.start = time.time()
     return self
   def __exit__(self, *args): 
     self.end = time.time()
     self.delta = self.end - self.start

def doTests(fracContiguousHyperVolumes = 0.01,numPointsPerSigma = 30):
    #Flag whether to generate/save plots
    doPlotPDFs = True
    doSavePlot = False

    if(doPlotPDFs or doSavePlot):
        import pylab as P
        i = 1

    #Define the sample size
    N = int(5e2)

    #Define the number of repetitions
    numRep = 5


    #set the random seed to a fixed number for repeatability
    random.seed(0)

    #Predeclare lists for the ISE and timings
    ISEMaster = []
    timingMaster = []


    for key,testObj in duongTestDict.iteritems():
        ISE = []
        timing = []

        #Repeat the test numRep times
        for n in range(numRep):

            #Draw a sample from this specific Duong test
            sample = testObj.getSample(N)

            #Calculate the self-consistent density estimate (and time it)
            with Timer() as t:
                pdfObj = sc.selfConsistentDensityEstimate(sample, \
                                                          numPointsPerSigma = numPointsPerSigma, \
                                                          fracContiguousHyperVolumes=fracContiguousHyperVolumes)

            #Get the transformed PDF
            scPDF = pdfObj.fSC

            #Get the axes
            x,y = tuple(pdfObj.xgrids)
            dx = x[1] - x[0]
            dy = y[1] - y[0]

            #Calculate the actual PDF at these axis points
            x2d,y2d = meshgrid(x,y)
            pdfStd = testObj(x2d,y2d)

            #Calculate the ISE
            ise = sum( (scPDF - pdfStd)**2 ) * dx * dy

            #Save the ISE and timing
            ISE.append(ise)
            timing.append(t.delta)

            if( (doPlotPDFs or doSavePlot) and n == 0):
                if(i == 1):
                    fig = P.figure(figsize=(6,8))
                plt = fig.add_subplot(3,2,i)
                #Plot the PDF standard 
                cs = plt.contour(x2d,y2d,pdfStd)
                #Plot the self consistent PDF at the same contour levels (and in semi-transparent black)
                plt.contour(x2d,y2d,scPDF,levels=cs.levels,color=[0,0,0,0.7])
                #Set the axis limits
                plt.set_xlim([-3,3])
                plt.set_ylim([-3,3])
                #Add a title marking which test
                plt.set_title('Target density {}'.format(key))
                #Increment the plot indexor
                i += 1

        #Append the ISE for each test to the master list
        ISEMaster.append(ISE)
        #Append the timing for each test to the master list
        timingMaster.append(timing)

    if(doPlotPDFs or doSavePlot):
        P.tight_layout()
    if(doSavePlot):
        #Save the figure to a file
        P.savefig("duongTests.png")

    if(doPlotPDFs):
        #Present the figure in a window
        P.show()

    print "#Test\tISE [arb. units]\ttime [s] ({} repetitions)".format(numRep)
    for ISE,timing,key in zip(ISEMaster,timingMaster,duongTestDict):
        #Calculate the ISE and timing statistics
        ISEmean = average(ISE)
        ISEstd = std(ISE)
        timingMean = average(timing)
        timingStd = std(timing)

        #Print a table of ISE/timing statistics
        print "{}\t{:3.2e} +/- {:3.2e} \t{:3.3f} +/- {:3.3f}".format(key,ISEmean,ISEstd,timingMean,timingStd)
```

In this last test, I modify the ``numPointsPerSigma`` parameter, which controls the grid spacing of the PDF estimate.  In the current implementation, the code sets the grid spacing for each dimension such that there will be ``numPointsPerSigma`` points resolving the widest part of the PDF.  The code sets the extent of the grid such that the min/max of the data values fall within the grid; hence numPointsPerSigma effectively controls both the number of grid points and the coarseness of the grid. So ``numPointsPerSigma = 10`` coarsens the grid resolution in real and fourier space, relative to ``numPointsPerSigma = 30``.  The previous implementation of the code led to parts of the ECF to 'bleeding' together so that the filter effectively included higher frequency components.  In the current version, where the axis spacings are now uncoupled, this does not seem to be as big of an issue.

In the previous version, coarsening the grid reduced the error; in this version, it worsens it slightly.  This seems like a reasonable property, since higher resolution should lead to a more accurate calculation of the ECF.  Regardless, the increase in error is tiny.
