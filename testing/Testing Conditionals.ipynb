#Defining a conditional standard

In the following we use sympy to derive the form of a bivariate contidional; we use this as a standard to test the conditional routine in selfConsistentDensityEstimate.

```python
import sympy as sp
sp.init_printing()
```

```python
#Define a bivariate normal function
def norm2d(x,y,mux=0,muy=0,sx=1,sy=1,r=0):
  coef = 1./(2*pi*sx*sy*sqrt(1.-r**2))
  expArg = -(1./(2*(1-r**2)))*( (x-mux)**2/sx**2 + (y-muy)**2/sy**2 - 2*r*  (x-mux)*(y-muy)/(sx*sy))
  return coef*exp(expArg)

def covMat(sx,sy,r):
  return [[sx**2,r*sx*sy],[r*sx*sy,sy**2]]


gausParams = [0.0,0.0,1.0,1.0,-0.9] #Standard, correlated bivariate
#Calculate the covariance matrix
gCovMat = covMat(*gausParams[2:])
mu = array(gausParams[:2])
```

```python
mux,muy,rho,x,y = sp.symbols(r'\mu_x \mu_y \rho x y',real=True)
sx,sy = sp.symbols(r'\sigma_x \sigma_y ',real=True,positive=True)

bivariateNormal = (1/(2*sp.pi*sx*sy*sp.sqrt(1-rho**2)))*sp.exp( \
                -(1/(2*(1-rho**2)))*( (x-mux)**2/sx**2 + (y-muy)**2/sy**2 - 2*rho*  (x-mux)*(y-muy)/(sx*sy)))

xMarginal = (1/(sx*sp.sqrt(2*sp.pi)))*sp.exp(-(x-mux)**2/(2*sx**2))
yMarginal = (1/(sy*sp.sqrt(2*sp.pi)))*sp.exp(-(y-muy)**2/(2*sy**2))

xConditional = bivariateNormal/yMarginal
yConditional = bivariateNormal/xMarginal
```

```python
#sp.Integral(xConditional,(x,-sp.oo,sp.oo)).doit()
yConditional.subs(rho,0).simplify()
```

```python
from sympy.utilities.lambdify import lambdastr,lambdify
conditionalFuncStr = lambdastr((x,y,mux,muy,sx,sy,rho),xConditional)
conditionalFunc = eval(conditionalFuncStr.replace('\\',""))
```

#Testing the marginal routine

```python
#Now estimate the coupla of samples drawn from the normal distribution

#Seed with 0 so results are repeatable
random.seed(0)
numSamples = 1e4
#Draw samples from the multivariate normal distribution
bivariateSamples = random.multivariate_normal(mu,gCovMat,(numSamples,)).transpose()

_pdf = sc.selfConsistentDensityEstimate(bivariateSamples,numPoints=513,fracContiguousHyperVolumes=1,positiveShift=True)
axes = _pdf.axes
conditional = _pdf.estimateConditionals(1,data = bivariateSamples,peakFrac = 0.0)

PP.contourf(axes[0],axes[1],_pdf.pdf,levels=pdfplot.levels)
PP.colorbar()
PP.show()

PP.contourf(axes[0],axes[1],conditional,levels=copplot.levels)
conditionalStandard = conditionalFunc(xg,yg,*gausParams)
PP.contour(axes[0],axes[1],conditionalStandard,alpha=0.8,linewidth=2)
PP.colorbar()
PP.xlim([-5,5])
PP.ylim([-5,5])
PP.show()
```

```python
#Now lets recalculate the conditional standard on the same grid as the estimate
xg,yg = meshgrid(axes[0],axes[1])
conditionalStandard = conditionalFunc(xg,yg,*gausParams)

igood = nonzero(logical_and( \
                logical_not(conditional.mask.ravel()), \
                conditionalStandard.ravel() > ma.amin(conditional)))[0]

oneToOne = linspace(amin(conditionalStandard.ravel()[igood]),amax(conditionalStandard.ravel()[igood]),128)
PP.loglog(conditionalStandard.ravel()[igood],conditional.ravel()[igood],'k.',alpha=0.1)
PP.loglog(oneToOne,oneToOne,'r-',linewidth=3,alpha=0.8)
PP.xlabel("Conditional PDF standard")
PP.ylabel("Conditional PDF estimate")
PP.show()
```

So it looks like a good portion of the estimated copula points fall on the 1:1 line.  However, there are a large fraction in the low part of the copula where the estimate is completely wrong (and too high).  This makes sense in a way, since this is the part of the PDF that will be systematically undersampled (and so the PDF and/or marginal estimates are likely to be close to 0 or negative).

Either way, I think that this demonstrates that the copula estimate technique that we're using here does a reasonable job of reproducing the the core part of the copula.
